### Задание 5

##### На языке Python предложить алгоритм вычисляющий численно предел с точностью  $\lim_{n\rightarrow + \infty} \frac{n}{n! ^{\frac{1}{n}}} $ с точностью $ \epsilon = 10^{-7}$


т.к. ищем предел для плюс бесконечности, беру несколько n с крупными возрастающими значениями. т.к. стандартные модули не справляются, не могут конвертировать столько крупное значение в float, подключаю либу sympy

```python
from sympy import *
from sympy.functions.combinatorial.factorials import factorial
import mpmath
```

```python
# к сожалению, как автоматизировать начальное приближение я пока не додумался
n = [100000, 100000000000000, 10000000000000000]
rslt = []
# пишу функцию поиска лимита конкретной функции до 7 знака
def lmt(n):
    #  перебираю значения заданного начального приближения
    for i in n:
        # складываю в список результатов
        rslt.append(i / (mpmath.power(mpmath.factorial(i),(1/i))))
    # далее я буду сравнивать значения резульатов между собой на предмет равенства на 7м знаке
    #  т.к. последний перебор будет за пределами индекса, оставлю пропуск ошибки
    try:
        for el in range(len(rslt)):
            if round(rslt[el],7) == round(rslt[el+1],7):
                print(f'найден, составляет {rslt[el]}')
            else:
                print('увеличьте начальное приближение')
    except IndexError:
        pass
# запускаю функцию
lmt(n)
```

### Задание 6

##### Предложить оптимизацию алгоритма, полученного в задании 5, ускоряющую его сходимость.

воспользоваться переменной как символов согласно функционалу предлагаемому либо sympy и функцией limit не получается ( через n = symbols('n')  ), т.к. выпадает ошибка при ппопытке извлечь корень, поэтому задаю приближение вручную, но использую встроенную функцию поиска предела

```python
n,k = 1000, 10000
def lmt2(n,k):
    rslt_n = limit(n/mpmath.root(factorial(n), n), oo, 0)
    rslt_k = limit(k/mpmath.root(factorial(k), k), oo, 0)
    if round(rslt_n, 7) == round(rslt_k, 7):
        print(f'найден, составляет {rslt_n}')
    else:
        print('увеличьте начальное приближение')
lmt2(n,k)
```
