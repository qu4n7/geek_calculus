область определения - также все множество вещественных чисел от $-\infty$ до $+\infty$

чтобы определить нули функции и их кратность надо найти кори уравнения:

$ x ^ 3 - x ^ 2 = 0 $

$ x^2(x-1) = 0 $

$ x_1, x_2 = 0, x_3 = 1 $

нуль x = 0 имеет кратность 2

нуль x = 1 имеет кратность 1

Отрезки знакопостоянства:

при x > 1 функция положительна

при x < 1 функция отрицательна

Интервалы монотонности:

при x > 1 функция монотонно возрастает

при x < 1 функция монотонно убывает

Четность функции:

функция нечетная, т.к. функция меняет значение на противоположное при изменении знака независимой переменной при этом график её симметричен относительно центра координат

функция не ограничена и не облаадет периодом

##### Найти предел: $lim\frac{3x^3-2x^2}{4x^2} $ при $x \rightarrow 0$

```python
# попробую применить близкие к 0 значения
x1, x2, x3 = 1e-8, 1e-13, 1e-17
y1 = (3*x1**3 - 2*x1**2) / (4*x1**2)
y2 = (3*x2**3 - 2*x2**2) / (4*x2**2)
y3 = (3*x3**3 - 2*x3**2) / (4*x3**2)
print(y1,y2,y3)
```

```python
# из расчеты выше предел определяется как -0.5
# проверяю либой
x = Symbol('x')
limit((3*x**3 - 2*x**2)/(4*x**2), x, 0)
```

предел: - 0.5

##### Найти предел: $lim\frac{\sqrt{1+x}-1}{\sqrt[3]{1+x}-1} $ при $x \rightarrow 0$

```python
# попробую применить близкие к 0 значения
x1, x2, x3 = 1e-8, 1e-13, 1e-17
y1 = ((1 + x1) ** 0.5 - 1) / ((1 + x1) ** (1/3) - 1)
y2 = ((1 + x2) ** 0.5 - 1) / ((1 + x2) ** (1/3) - 1)
#y3 = ((1 + x3) ** 0.5 - 1) / ((1 + x3) ** (1/3) - 1)
print(y1,y2)
# скорее всего, 1.5
```

```python
x = Symbol('x')
limit(((1 + x) ** 0.5 - 1) / ((1 + x) ** (1/3) - 1), x, 0)
```

предел: 1.5

##### Найти предел: $lim(\frac{x + 3}{x})^{4x+1} $ при $x \rightarrow \infty$

```python
# попробую применить близкие к oo значения
x1, x2, x3 = 1e+13, 1e+17, 1e+23
y1 = ((x1+3) / x1) ** (4*x1+1)
y2 = ((x2+3) / x2) ** (4*x2+1)
y3 = ((x3+3) / x3) ** (4*x3+1)
print(y1,y2,y3)
# скорее всего, 1
```

```python
x = Symbol('x')
limit(((x+3) / x1) ** (4*x+1), x, oo)
```

предел: oo

##### Найти предел: $lim(\frac{sin(2x)}{4x}) $ при $x \rightarrow 0$

```python
# попробую применить близкие к 0 значения
x1, x2, x3 = 1e-8, 1e-13, 1e-17
y1 = (np.sin(2*x1)) / (4*x1)
y2 = (np.sin(2*x2)) / (4*x2)
y3 = (np.sin(2*x3)) / (4*x3)
print(y1,y2,y3)
# скорее всего, 0.5
```

предел: 0.5
